# Text normalisation on Review Text

```python
def clean_string(s):
    r1 = re.sub('[^A-Za-z]+',' ',str(s))
    r1 = r1.strip().lower()
 
 # removing stopwords
    no_stp_wrd = []
    if r1 not in stopwords.words():
        no_stp_wrd.append(r1)
 
 # performing stemming
    stemmer = PorterStemmer()
    final_list = []
    for i in no_stp_wrd:
        final_list.append(stemmer.stem(str(i)))
        
    return final_list
```

# Sentiment Analysis on Filtered Review column

```python
sia = SentimentIntensityAnalyzer()
num_records = reviews_df.shape[0]
```

```python
num_records
```

```python
ones_array = np.ones(num_records)
```

```python
ones_array
```

```python
reviews_df['Compound Score'] = ones_array
reviews_df.head()
```

```python
reviews_df['Filtered Review']
```

```python
len(reviews_df['Filtered Review'])
```

```python
for i in range(len(reviews_df['Filtered Review'])):
    element = reviews_df['Filtered Review'].iloc[i][0]
#    print(element)
#    break
    score = sia.polarity_scores(element)
#     print(score)
    compound = score['compound']
    reviews_df['Compound Score'].iloc[i] = compound
```

```python
reviews_df.head()
```

```python
category = []
for i in reviews_df['Compound Score']:
    if i > 0:
        category.append("Positive")
    elif i < 0:
        category.append("Negative")
    else:
        category.append("Neutral")
        
reviews_df['Sentiment_Category'] = category
```

# Applying Naive Bayes Classification Algorithm

```python
reviews_df.head()
```

```python
X = reviews_df['Sentiment_Category']

y = reviews_df['Recommended IND']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_test
```

```python
count_vect = CountVectorizer()
tfidf_tfm = TfidfTransformer()
train_df_counts = count_vect.fit_transform(X_train)
train_df_tfidf = tfidf_tfm.fit_transform(train_df_counts)
```

```python
clf = MultinomialNB().fit(train_df_tfidf,y_train)
```

```python
print(train_df_tfidf)
```

```python
count_test_vect = count_vect.transform(X_test)
y_predict = clf.predict(count_test_vect)
```

```python
print(count_test_vect)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
print(cm)
```

```python
print("True Negative",cm[0][0])
print("True Positive",cm[1][1])
print("False Negative",cm[0][1])
print("Total",sum(sum(cm)))
print('Accuracy',(cm[0][0]+cm[1][1])/sum(sum(cm)))
```
